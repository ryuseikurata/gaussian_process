```julia
using Distributions
using Plots

include("gaussian_process.jl")
include("gaussian_kernel.jl")

gp = GaussianProcess(GaussianKernel(1))
xs = collect(-4:0.1:4)
Plots.plot(xs, rand(dist(gp, xs), 5),
    label = "", xlabel = "x", ylabel = "f",
    linewidth = 2,
    title = "" )
```

```julia
using Distributions
using Plots

include("gaussian_process.jl")
include("gaussian_kernel.jl")

# ノイズ項を追加する
gp = GaussianProcess(GaussianKernel(1), 0.01)
xs = collect(-4:0.01:4)
Plots.plot(xs, rand(dist(gp, xs), 5),
    label = "", xlabel = "x", ylabel = "f",
    linewidth = 2,
    title = "")
```
